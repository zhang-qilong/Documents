```python
#字符串处理方法（str.function）均为创建副本，不改变字符串
a = 'ab cd ef'
b = 'EF GH IJ'
newstr = ' 123 321 '
print(a.title())  # 首字母大写
print(a.upper())  # 全部大写
print(b.lower())  # 全部小写
a.title()
print(a, b)  # 不改变字符串的值
c, d, e = a.split()  # 分割字符串，双引号指定按什么分隔（默认为空格），返回值为list
print('{}\n{}\n{}'.format(c, d, e))  # 默认按照空格切分
print(a.count('a'))  # 统计指定字符串出现的次数
print(a.replace('a', '1'))  # 替换字符串中字符
print(a.center(25, '*'))  # 居中显示（长度，填充符号）
print(newstr.strip())
print(newstr.strip(' 1'))  # 去掉字符串首尾指定字符，默认为空格
print(newstr.lstrip())
print(newstr.rstrip())  # 左、右（left、right）

print('*'.join(a))  # 在每个字符之间加上指定str
```

```python
#凯撒密码
#将字母表中的元素循环替换成其后第三个字符
password = input('请输入加密文本：')
for i in password:
   if (ord(i) >= 65) & (ord(i) <= 88):
       print('原文本为：', chr(ord(i)+3), end='')
       # end表示将自动换行替换成指定内容
   elif (ord(i) > 88) and (ord(i) <= 91):
       print('原文本为：', chr(ord(i)+3-26), end='')
   else:
       print('非英文文本！')
```

```python
#函数与代码复用
def f(a, b=20):  # 指定函数默认参数，a为非可选参数
   return a+b
print(f(2,2))
print(f(2))
print(f(b=2, a=5))  # 根据形参名称传递
```

```python
#全局变量在函数内部使用,globals声明
z = 2
def f(a,b):
   global z
   # z = 0
   return z*a*b
print(f(2,2))
```

```python
#集合set，不可索引，元素类型为不可变数据类型
#（集合中不能以集合为元素）
#可以使用{}或set()创建集合
#创建空集合必须用set()，{}创建空字典
s = {1010, 'hello', 98.5, 1000}
t = {1000, 1}
print(s)
print(type(s))
print(s-t)  # 差集
print(s&t)  # 交集
print(s^t)  # 补集，两集合非共同元素
print(s|t)  # 并集
#集合操作函数、方法
a = set({})  # 花括号可以不写
a.add(10)
a.add(11)
a.add(12)
print(a)  # 先添加再输出，直接输出显示为空集
print(len(a))
a.remove(10)
print(a)
a.clear()
print(a)
```

```python
# 映射类型（dict为主），字典中的键为无序不重复的
# 键必须为不可变数据类型
# 字典中指定key=value，若key不存在则创建
a = {1:'1', 2:'2', 3:'3'}  # 或a = dict()创建空字典
print(a.items())  # 输出所有键值对
print(a.get(1))  # 输出指定建对应值 dict.get(key, value), 若key不存在，返回value
b = a.pop(1)  # 弹出键对应键值对
print(a, b)
c = a.popitem()  # 随机取出键值对并删除,返回元组
print(a, c)
a.clear()  # 清除
```
